```python
import datetime as dt
from datetime import datetime
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# read csv file
data = pd.read_csv('US Tech_.csv')
data.head(264)
```

```python
dateFormatter = "%d/%m/%Y"
for i in range(264):
    datetime.strptime(data.Date[i], dateFormatter)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
for i in range(264):
    data['Dernier'][i]=data['Dernier'][i][0:5]
    data['Ouv.'][i]=data['Ouv.'][i][0:5]
    data['Plus Haut'][i]=data['Plus Haut'][i][0:5]
    data['Plus Bas'][i]=data['Plus Bas'][i][0:5]
```

```python
data['Dernier'] = data['Dernier'].astype('float')
data['Ouv.'] = data['Ouv.'].astype('float')
data['Plus Haut'] = data['Plus Haut'].astype('float')
data['Plus Bas'] = data['Plus Bas'].astype('float')
```

```python
data.head(264)
```

```python
data['Series'] = np.sort(np.arange(1,len(data)+1))[::-1]
```

```python
data
```

```python
# create 12 month moving average
data['MA12'] = data['Dernier'].rolling(12).mean()
data['MA10'] = data['Dernier'].rolling(10).mean()

# plot the data and MA
#xgridoff
fig = px.line(data, x="Series", y=["Dernier", "MA12","MA10"], template = 'plotly_white')
fig.show()
```

```python
# extract month and year from dates
for i in range(264):
    data['Day'][i]=data['Date'][i][0:2]
    data['Month'][i]=data['Date'][i][3:5]
```

```python
data
```

```python
# drop unnecessary columns and re-arrange
data.drop(['Date', 'Ouv.','Plus Haut','Plus Bas','Vol.','Variation %','MA12','MA10'], axis=1, inplace=True)
data = data[['Series', 'Dernier', 'Month', 'Day']] 
```

```python
data
```

```python
# check the head of the dataset
data.head()
```

```python
data.info()
```

```python
# split data into train-test set
train = data[data['Series'] < 201]
test = data[data['Series'] >= 201]
# check shape
train.shape, test.shape
```

```python
# import the regression module
from pycaret.regression import *
# initialize setup
s = setup(data = train, 
          test_data = test, 
          target = 'Dernier', 
          fold_strategy = 'timeseries', 
          numeric_features = ['Series'], 
          fold = 3, 
          transform_target = True, 
          session_id = 123)
```

```python
best = compare_models(sort = 'MAE')
```

```python
prediction_holdout = predict_model(best)
```

```python
# generate predictions on the original dataset
predictions = predict_model(best, data=data)
```

```python
# add a vertical rectange for test-set separation
fig.add_vrect(x0="1960-01-01", x1="1960-12-01", fillcolor="grey", opacity=0.25, line_width=0)
fig.show()
```

```python
final_best = finalize_model(best)
```

```python
future_dates = pd.date_range(start = '1961-01-01', end = '1965-01-01', freq = 'MS')
future_df = pd.DataFrame()
future_df['Month_'] = [i.month for i in future_dates]
future_df['Year'] = [i.year for i in future_dates]    
future_df['Series'] = np.arange(145,(145+len(future_dates)))
future_df.head()
```

```python
predictions_future = predict_model(final_best, data=future_df)
predictions_future.head()
```

```python
concat_df = pd.concat([data,predictions_future], axis=0)
concat_df_i = pd.date_range(start='1949-01-01', end = '1965-01-01', freq = 'MS')
concat_df.set_index(concat_df_i, inplace=True)
fig = px.line(concat_df, x=concat_df.index, y=["#Passengers", "Label"], template = 'plotly_white')
fig.show()
```
